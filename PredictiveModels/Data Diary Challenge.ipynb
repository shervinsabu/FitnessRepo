```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
df2 = pd.DataFrame(scaler.fit_transform(x))
df2
```

```python
x_trans = scaler.fit_transform(x)
x_trans
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_trans, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred,average='weighted')
```

```python
from sklearn.metrics import f1_score,recall_score
recall_score(y_test,y_pred,average='weighted')
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred))
disp.plot()
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
gau = GaussianNB()
gau.fit(x_train,y_train)
y_pred = gau.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from scipy import interp
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.linear_model import LogisticRegression
# Learn to predict each class against the other
yhat = gau.predict_proba(x_test)
y_test_bin = label_binarize(y_test,classes=np.unique(y_test))
n_classes = len(np.unique(y_test)) # number of class
# Compute ROC curve and ROC area for each class
fpr = {}
tpr = {}
roc_auc = dict()
thresh = {}
for i in range(n_classes):
    fpr[i], tpr[i], thresh[i] = roc_curve(y_test_bin[:, i], yhat[:, i],drop_intermediate=False)
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.figure()
    lw = 2 # line_width
    plt.plot(fpr[i], tpr[i], color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[i]) 

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('MultiClass ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
